```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('dataset_lab_3_2.csv') 
dataset
```

```python
X = dataset.iloc[:, 1:-1].values 
y = dataset.iloc[:, -1].values 
```

```python
from sklearn.linear_model import LinearRegression 
lin_reg = LinearRegression() 
lin_reg.fit(X, y) 
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
from sklearn.preprocessing import PolynomialFeatures 
poly_regr = PolynomialFeatures(degree = 4) 
X_poly = poly_regr.fit_transform(X) 
lin_reg_2 = LinearRegression() 
lin_reg_2.fit(X_poly, y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue') 
plt.title('Truth or Bluff (Linear Regression)') 
plt.xlabel('Position Level') 
plt.ylabel('Salary') 
plt.show() 
```

```python
plt.scatter(X, y, color = 'red') 
plt.plot(X, lin_reg_2.predict(poly_regr.fit_transform(X)), color = 'blue') 
plt.title('Truth or Bluff (Polynomial Regression)') 
plt.xlabel('Position level') 
plt.ylabel('Salary') 
plt.show() 
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1)) 
plt.scatter(X, y, color = 'red') 
plt.plot(X_grid, lin_reg_2.predict(poly_regr.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)') 
plt.xlabel('Position level')
plt.ylabel('Salary') 
plt.show()
```
