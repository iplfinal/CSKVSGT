```python
import random
from matplotlib import pyplot as pltplt
import math
```

## Part1

```python
d=0
D = []
avg=0
i = 0
while (avg<=4):
    x=0
    y=0
    for j in range(i):
        dx,dy = random.choice([(0,-1),(0,1),(-1,0),(1,0)])
        x += dx
        y += dy
    d = math.sqrt((x**2)+(y**2))
    D.append(d)
    avg = sum(D)/len(D)
    i+=1
print(i)
```

## Part 2

```python
A1=0
A2=0
X1 = []
Y1 = []
X2 = []
Y2 = []
for i in range(20000):
    x = random.uniform(-1.00,1.00)
    y = random.uniform(-1.00,1.00)
    if((x**2)+(y**2)>1):
        A1+=1
        X1.append(x)
        Y1.append(y)
    else:
        A2+=1
        X2.append(x)
        Y2.append(y)

print("Actual value: ",(3.1415/4))
print("Predicted value: ",A2/(A1+A2))
plt.scatter(X1, Y1, color="green")
plt.scatter(X2, Y2, color="orange")
plt.show()
```
